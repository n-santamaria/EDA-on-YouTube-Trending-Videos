# **7. Data Pre-Processing**
---

- Here we will perform **Data Preprocessing** on our dataset, to make the data usable for **EDA**.


```python
trending_us_videos.head()
```

### Dropping columns:-


- ***items__kind*** and ***items__snippet__channelId*** are constant value, we will drop these columns.


- ***description***, ***thumbnail_link***, ***tags***, ***etag***, ***kind***, and ***items__etag*** does not provide any insight, so we will drop these columns too.

```python
# Dropping columns which doesn't add value to this EDA

trending_us_videos.drop(['description', 'thumbnail_link', 'tags', 'items__kind', 
                         'items__snippet__channelId', 'etag', 'kind', 'items__etag'], axis=1, inplace=True)
print('Successfully dropped irrelevant columns.');
```

### Changing columns datatype:-

- ***trending_date*** and ***publish_time*** are categorical, we will convert them to datetime format.


```python
# Converting trending_date, publish_time column values from categorical/object to datetime

trending_us_videos['trending_date'] = pd.to_datetime(trending_us_videos['trending_date'], format='%y.%d.%m')
trending_us_videos['publish_time'] = pd.to_datetime(trending_us_videos['publish_time'], format='%Y-%m-%dT%H:%M:%S.%f')
display(trending_us_videos.info())
```

### Changing columns datatype:-

- There're total ***48*** rows which needs to be ***dropped/removed***.

```python
# Removing/dropping all 48 duplicated rows.

trending_us_videos.drop_duplicates(inplace=True)
print('Successfully dropped duplicate rows.')

# After dropping rows we need to reset the index.
trending_us_videos.reset_index(drop=True, inplace=True)
print('Successfully reseted index.')
```

```python
trending_us_videos.head()
```

### video_id and title aren't matching:-

- As we understand each and every video should be having a unique ***video_id*** and a ***title***.
- But we have noticed that the ***unique video_id = 6351*** whereas ***unique title = 6455***. 
- There's a difference of ***109*** count.
- In the below code, we got to know that there're changes in the title of the video after it was trended for the 1st time.

```python
print('No. of videos ids: ', trending_us_videos['video_id'].nunique())
print('Np. of video titles: ', trending_us_videos['title'].nunique())

vt_change = trending_us_videos.groupby('video_id')['title'].nunique()
vt_change = vt_change[vt_change != 1].keys().tolist()
vt_change =  trending_us_videos[trending_us_videos['video_id'].isin(vt_change)][['video_id','title']]
vt_change.drop_duplicates(inplace=True)
vt_change.reset_index(drop=True, inplace=True)
vt_change
```

```python
profile = trending_us_videos.profile_report(title="Trending US YouTube Videos - Profiling after Data Pre-Processing", progress_bar=False, minimal=True)
profile.to_file(output_file="profiling_after_preprocessing.html")
print('Profiling successfully completed.')
```

# **9. Exploratory Data Analysis**

### Common Configuration

```python
def plot_pre_config(figsize):
    sns.set_style("whitegrid")
    bar, ax = plt.subplots(figsize=figsize)
    return ax

def plot_post_config(title, xlabel, ylabel, fontsize, labelsize):
    ax.tick_params(labelsize=labelsize)
    ax.set_title(title, fontsize=fontsize, fontweight='bold')
    ax.set_xlabel(xlabel, fontsize=fontsize, fontweight='bold')
    ax.set_ylabel(ylabel, fontsize=fontsize, fontweight='bold')

month_lookup = {
    1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 
    9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'
}

days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

print('Configuration loaded successfully.')
```

### 1. Here we find out the total number trending videos: Nov, Dec - 2017 and Jan to June - 2018.

```python
trending_by_date = trending_us_videos.groupby(trending_us_videos['trending_date'].dt.year)['video_id'].count()
plot_pre_config((8,8))
ax = sns.barplot(x=trending_by_date.index, y=trending_by_date, ci=None, palette='muted', orient='v')
for rect in ax.patches:
    ax.text(rect.get_x() + rect.get_width()/2., rect.get_height(), int(rect.get_height()), fontsize=18)
plot_post_config("Trending videos in 2017 and 2018", "Year", "No. of Videos", 18, 20)
```

### Observation:-
- **Nov, Dec - 2017** have 9600 trending videos.<br>
- **Jan to June - 2018** have 31301 trending videos.<br>

### 2. Here we find out the total number trending videos in a month.

```python
trending_by_month = trending_us_videos.groupby([trending_us_videos['trending_date'].dt.year, trending_us_videos['trending_date'].dt.month])['video_id'].count()
cols = []
for i,j in trending_by_month.items():
    cols.append(str(month_lookup[i[1]]) +'-'+str(i[0]))
plot_pre_config((15,10))
ax = sns.barplot(x=cols, y=trending_by_month, ci=None, palette='muted', orient='v')
for rect in ax.patches:
    ax.text(rect.get_x() + rect.get_width()/2., rect.get_height(), int(rect.get_height()), fontsize=15)
plot_post_config("Trending videos in 2017 and 2018", "Month-Year", "No. of Videos", 18, 15)
```

### Observation:-
- **December-2017, March-2018, and May-2018** are the top 3 months having most of the trending videos.

### 3. Here we find out number of videos published by days of week.

```python
trending_by_day = trending_us_videos.groupby(trending_us_videos['publish_time'].dt.day_name()).count().reindex(days) 
plot_pre_config((20,10))
ax = sns.barplot(x=trending_by_day.index, y=trending_by_day.video_id, ci=None, palette='muted', orient='v')
for rect in ax.patches:
    ax.text(rect.get_x() + rect.get_width()/2., rect.get_height(), int(rect.get_height()), fontsize=18)
plot_post_config("Published videos by days of week", "Days of Week", "No. of Videos", 18, 20)
```

### Observation:-
- Most of the videos were published during **weekdays**.

### 4. Here we find out number of videos published by date.

```python
trending_by_date = trending_us_videos.groupby(trending_us_videos['publish_time'].dt.day)['video_id'].count()
trending_by_date
plot_pre_config((20,10))
ax = sns.lineplot(x=trending_by_date.index, y=trending_by_date)
ax.set_xlim(1,31)
ax.set_xticks(range(1,32))
plot_post_config("Published videos by date", "Date", "No. of Videos", 18, 15)
```

### Observation:-
- Most of the videos where published on **20th** followed by **21st** and **22nd**.

### 5. Here find out the number of videos published by hour.

```python
trending_by_hour = trending_us_videos.groupby(trending_us_videos['publish_time'].dt.hour)['video_id'].count()
plot_pre_config((20,10))
ax = sns.lineplot(x=trending_by_hour.index, y=trending_by_hour)
ax.set_xlim(1,24)
ax.set_xticks(range(1,25))
plot_post_config("Published videos by hour", "Hour", "No. of Videos", 18, 15)
```

### Observation:-

### 14. Here we find out how long does it take for a video to trend after been published.

```python
diff = trending_us_videos[['video_id','title','publish_time','trending_date']].sort_values(by=['video_id','trending_date'])
diff = diff.groupby('video_id').first()
diff['publish_time'] = diff['publish_time'].dt.tz_localize(None)
diff['diff_publish_trending'] = (diff['trending_date'] - diff['publish_time']).dt.days
diff = diff.sort_values(by='diff_publish_trending')
diff.loc[diff['diff_publish_trending'] == -1, ['diff_publish_trending']] = 0
diff = diff['diff_publish_trending'].value_counts().sort_values(ascending=False)[:21]
plot_pre_config((15,10))
ax = sns.lineplot(x=diff.index, y=diff.values)
plot_post_config('Difference between Publish and Trending Date','No. of Days Difference','No. of Videos',16, 12)
```

### Observation:-
- We can see that most of the published videos were trended within 5 to 10 days of published date.
